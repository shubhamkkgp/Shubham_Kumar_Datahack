```python
data = pd.merge(df_train,df_test)
```

```python
df_train_label = pd.read_csv('dataset and all/training_set_labels.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

feature_columns = df_train.columns.tolist()

target_columns = ['xyz_vaccine', 'seasonal_vaccine']

X = df_train[feature_columns]
y_xyz = df_train_label['xyz_vaccine']
y_seasonal = df_train_label['seasonal_vaccine']

numerical_features = X.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
X_preprocessed = preprocessor.fit_transform(X)

X_train, X_val, y_xyz_train, y_xyz_val = train_test_split(X_preprocessed, y_xyz, test_size=0.2, random_state=42)
X_train, X_val, y_seasonal_train, y_seasonal_val = train_test_split(X_preprocessed, y_seasonal, test_size=0.2, random_state=42)

xyz_model = RandomForestClassifier(random_state=42)
xyz_model.fit(X_train, y_xyz_train)

seasonal_model = LogisticRegression(random_state=42, max_iter=1000)
seasonal_model.fit(X_train, y_seasonal_train)

y_xyz_pred = xyz_model.predict_proba(X_val)[:, 1]
y_seasonal_pred = seasonal_model.predict_proba(X_val)[:, 1]

xyz_auc = roc_auc_score(y_xyz_val, y_xyz_pred)
seasonal_auc = roc_auc_score(y_seasonal_val, y_seasonal_pred)

print(f'XYZ Vaccine Model AUC: {xyz_auc}')
print(f'Seasonal Vaccine Model AUC: {seasonal_auc}')

X_test = df_test[feature_columns]
X_test_preprocessed = preprocessor.transform(X_test)

xyz_test_pred = xyz_model.predict_proba(X_test_preprocessed)[:, 1]
seasonal_test_pred = seasonal_model.predict_proba(X_test_preprocessed)[:, 1]

submission = pd.DataFrame({
    'respondent_id': df_test['respondent_id'],
    'xyz_vaccine': xyz_test_pred,
    'seasonal_vaccine': seasonal_test_pred
})

submission.to_csv('submission.csv', index=False)
```
